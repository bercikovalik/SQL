```sql
SELECT MIN(NumberOfRooms) AS legkisebb_szoba_méret, 
       MAX(NumberOfRooms) AS legnagyobb_szoba_méret, 
       AVG(NumberOfRooms) AS atlagos_szoba_méret
FROM Flat
```
